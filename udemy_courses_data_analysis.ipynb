```python
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sb
```

```python
# Distribution of subjects

plt.figure(figsize=(10, 6))
sb.countplot(data=df, x='subject', hue = "subject")
plt.title('Distribution of Subjects')
plt.xlabel('Subject')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of course levels

plt.figure(figsize=(8, 6))
sb.countplot(data=df, x='level', hue = "level")
plt.title('Distribution of Course Levels')
plt.xlabel('Level')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation between number of subscribers and number of reviews

plt.figure(figsize=(8, 6))
sb.scatterplot(data=df, x='num_subscribers', y='num_reviews')
plt.title('Correlation between Number of Subscribers and Number of Reviews')
plt.xlabel('Number of Subscribers')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
# Distribution of prices

plt.figure(figsize=(20, 6))
sb.histplot(data=df, x='price', bins=10, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Relationship between content duration and number of lectures

plt.figure(figsize=(10, 6))
sb.barplot(data=df, x='level', y='content_duration', hue = "level")
plt.title('Content Duration Based on Number of Lectures in Each Level')
plt.xlabel('Level')
plt.ylabel('Content Duration (hours)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of paid vs. free courses

plt.figure(figsize=(6, 4))
df['is_paid'].value_counts().plot(kind='bar')
plt.title('Distribution of Paid vs. Free Courses')
plt.xlabel('Is Paid?')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
